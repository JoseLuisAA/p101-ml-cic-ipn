# Análisis para características categóricas

```python
c1 = titanic['Survived']
#c1.value_counts()
(c1.value_counts(normalize=True)*100).round(2)
```

```python
c1.value_counts().plot.bar()
```

```python
c1.value_counts(normalize=True).plot.pie(autopct="%.1f%%")
```

```python
c2 = titanic['Pclass']
c2.value_counts(normalize=True)
```

```python
c3 = titanic['Sex']
c3.value_counts(normalize=True)
```

```python
c3.value_counts().plot.pie(autopct="%.1f%%",explode=[0.1,0])
```

```python
c4 = titanic['Age']
c4.fillna(-1).astype(int).value_counts()
```
