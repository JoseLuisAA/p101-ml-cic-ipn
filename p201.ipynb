# Condicionales en características categóricas

### Respuesta de la IA

```python

# Suponiendo que tu DataFrame se llama df y tiene la columna 'Age'
# Definimos los límites (bins) y las etiquetas (labels)
bins = [0, 12, 20, 39, 64, 150] # El 150 es un límite superior seguro
labels = ['child', 'young', 'adult', 'sinior', 'old']

# Creamos la nueva columna c10
c10 = pd.cut(titanic['Age'], bins=bins, labels=labels, right=True)
c10.value_counts(dropna=False)
```

### Uso de .map() [.apply()]

```python
def clasificar_edad(edad):
    if edad <= 12:
        return 'child'
    elif edad <= 20:
        return 'young'
    elif edad < 40: 
        return 'adult'
    elif edad < 65:
        return 'sinior'
    elif edad >= 65:
        return 'old'
    else:
        return 'unknown' 

c10_1 = titanic['Age'].apply(clasificar_edad)
c10_1.value_counts(dropna=False)
```

### Gráficas

```python
c10.value_counts().reindex(['child', 'young', 'adult', 'sinior', 'old']).plot.pie(autopct="%.1f%%")
```

## Respuesta del alumno

```python
titanic['Family'] = c6 + c5

def familia(x):
    if x <= 0:
        return '0'
    elif x == 1:
        return '1'
    elif x >= 2:
        return '2+'

c11 = titanic['Family'].apply(familia)
c11.value_counts(dropna=False)
```

## Respuesta de la IA

```python
titanic['family_size'] = titanic['SibSp'] + titanic['Parch']

# Definimos los límites: 
# -1 a 0 -> '0'
# 0 a 1  -> '1'
# 1 a 100 -> '2+'
titanic['c11'] = pd.cut(titanic['family_size'], 
                   bins=[-1, 0, 1, 100], 
                   labels=['0', '1', '2+'])
titanic['c11'].value_counts(dropna=False)
```

```python
c11.value_counts().plot.pie(autopct="%.1f%%", explode=[0.1, 0, 0])
```

### ¿Cuál es la probabilidad de que un hombre sobreviva?

```python
sobrevivientes = {}
sexos = by_sex.unique()
for sexo in sexos:
    cont = titanic[by_sex == sexo].shape[0]
    sobrevivientes[sexo] = cont
print(sobrevivientes)
```

```python
total = {}
sexos = titanic['Sex'].unique()
for sexo in sexos:
    cont = titanic[(titanic['Sex'] == sexo)].shape[0]
    total[sexo] = cont
print(total)
```

```python
man_surv = sobrevivientes["male-1"] / titanic.shape[0] * 100
print(f'Probabilidad de supervivencia para hombres: {man_surv:.2f}')
```

### ¿Cuál es la probabilidad de que siendo hombre el pasajero sobreviva?

```python
man_man = sobrevivientes["male-alive"] / total["male"] * 100
print(f'Probabilidad de supervivencia entre hombres: { man_man:.2f}')
```

### ¿Cuál es la probabilidad de que una mujer sobreviva?

```python
woman_surv = sobrevivientes["female-alive"] / titanic.shape[0] * 100
print(f'Probabilidad de supervivencia para mujeres: {woman_surv:.2f}')
```

### ¿Cuál es la probabilidad de que siendo mujer el pasajero sobreviva?

```python
woman_woman = sobrevivientes["female-alive"] / total["female"] * 100
print(f'Probabilidad de supervivencia entre mujeres: {woman_woman:.2f}')
```
